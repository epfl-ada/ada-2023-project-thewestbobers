# Trailblazers of Cinematic Genres

## first movie of each most popular genre


The inaugural films of any genre typically forge a path for future narratives, setting the stage with innovative themes, styles, or storytelling techniques. These films often stand as quintessential "pivot movies"—landmarks that inspire and influence the evolution of their respective genres. Bel(after the code)ow, we present a list of the pioneering films that represent the genesis of the thirty most celebrated genres in cinema history:

```python
movies['date'] = pd.to_datetime(movies['date'], errors='coerce')

# List of 30 most common genres
most_common_genres = most_common

# Initialize an empty list to store the earliest movies for each genre
earliest_movies_list = []

for genre in most_common_genres:
    # Filter for movies that include the genre and are not missing a date
    genre_movies = movies[movies['genres'].apply(lambda genres: genre in genres) & movies['date'].notnull()]
    
    # Find the earliest movie for the genre
    if not genre_movies.empty:
        earliest_movie_idx = genre_movies['date'].idxmin()
        earliest_movie = genre_movies.loc[earliest_movie_idx, ['name', 'date', 'genres']]
        # Append the earliest movie details to the list
        earliest_movies_list.append(earliest_movie)

# Convert the list of Series to a DataFrame
earliest_movies_per_genre = pd.DataFrame(earliest_movies_list)

# Reset the index of the resulting DataFrame
earliest_movies_per_genre.reset_index(drop=True, inplace=True)

# Now 'earliest_movies_per_genre' contains the earliest movie for each of the 30 genres
# with only the 'name', 'date', and 'genres' columns
for i in range(len(most_common)):
    print(   most_common[i] + ": \n" + str(earliest_movies_per_genre[['name', 'date', 'genres']].iloc[i] ))
```

```python
# Explode the 'genres' list into separate rows
df_exploded = df.explode('genres')

# Group by year and genre, then count occurrences
genre_counts = df_exploded.groupby(['year', 'genres']).size()

# Convert the Series into a DataFrame
genre_counts_df = genre_counts.reset_index(name='count')

# Calculate the total count for each year
total_counts = genre_counts_df.groupby('year')['count'].transform('sum')

# Calculate the percentage
genre_counts_df['percentage'] = (genre_counts_df['count'] / total_counts) * 100
```

```python

# Assuming genre_counts_df is your DataFrame

# Filter the DataFrame for the selected years
years_to_plot = [2010, 2009,2008]  # You can add more years to this list
data_to_plot = genre_counts_df[genre_counts_df['year'].isin(years_to_plot)]

# Determine the top 10 genres (based on frequency of appearance)
top_genres = data_to_plot['genres'].value_counts.head(15).index

# Filter the data to include only the top 10 genres
data_filtered = data_to_plot[data_to_plot['genres'].isin(top_genres)]

# Sort the filtered data by year and then by genres
data_sorted = data_filtered.sort_values(['year', 'genres'])

# Extract the unique genres and years
unique_genres = data_sorted['genres'].unique()
unique_years = data_sorted['year'].unique()

# Assign a unique color for each of the top 10 genres using a different colormap for more variety
genre_colors = {genre: plt.cm.tab20(i / len(unique_genres)) for i, genre in enumerate(unique_genres)}

# Initialize figure and axis
fig, ax = plt.subplots()

# Create a stacked bar chart
for year in unique_years:
    bottom = np.zeros(len(unique_years))  # Initialize the bottom array for this year
    year_data = data_sorted[data_sorted['year'] == year]

    for genre in unique_genres:
        genre_data = year_data[year_data['genres'] == genre]
        if not genre_data.empty:
            percentage = genre_data['percentage'].values[0]
            ax.bar(str(year), percentage, bottom=bottom, color=genre_colors[genre], label=genre)
            bottom += percentage

# Set y-axis limits to 0-100
ax.set_ylim(0, 10)

# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Percentage')
ax.set_title('Stacked Genre Percentages by Year for Top 10 Genres')

# Add a legend for genres
# Place the legend outside of the plot
ax.legend(genre_colors.keys(), title="Genres", bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to make room for the legend
plt.tight_layout()

# Show the plot
plt.show()
```
